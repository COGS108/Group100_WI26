## Research Question

**How do variables such as game genre, price, average length of playtime, and achievement number influence how Steam games are rated on the Steam shop globally since the platform's launch in 2003? Can these variables be used to predict how a Steam game might be rated?**

Note: All the variables are attributes of games listed in the Steam storefront page for a game.
Steam game genre refers to the available platforms and number of players it is targeting, price refers to the amount of currency in dollars required to purchase (including free games), average length of playtime refers to the average amount of time in hours it takes a player to reach an ending of the game for the first time, and achievement number refers to how many achievements can be attained by meeting in-game requirements for a game.
We are restricting our population to games on Steam, and thus only reviewing ratings left on the Steam platform. 
We are interested in sampling across a time period of 26 years and investigating whether there are any significant patterns in the aforementioned variables. 



## Background and Prior Work

Steam, a digital distribution service, lets its users buy video games from its digital store with a catalog of thousands of publishers, as well as install into and play video games from the users’ own digital library.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Important attributes of the games, such as their genre, price, number of game achievements are displayed on their store pages, allowing interested players to quickly get a feel for the game they may want to purchase.

#### Which genre is best?

We wanted to explore the effect that genre had on how well a game performed in the Steam shop. To achieve this, we used the ratio of positive reviews to total reviews as our test statistic and used one-hot encoding to compensate for the fact that each game has multiple genres attached. One challenge we encountered while creating this EDA was that the 'rating_percent' column is originally a string containing the code for a Python list that held the genres. To solve this, we had to import the 'ast' package. Additionally, some genres had artificially inflated scores due to having very few entries, so we dropped them. 

```python
# create a copy of the steam_games dataframe for manipulation
copy = steam_games.copy()

# add a column 'rating_percent' that describes the ratio of positive reviews to total reviews  
copy['rating_percent'] = copy['Positive'] / (copy['Positive'] + copy['Negative'])
# convert the 'genres' column from strings to lists 
copy['genres'] = copy['genres'].apply(lambda x: ast.literal_eval(x))

# get all the unique genres from the genres column
genres = pd.Series([i for lst in copy['genres'] for i in lst]).unique()

# create a new one-hot column for every unique genre in 'genres'
for genre in genres:
    copy[genre] = copy['genres'].apply(lambda x: genre in x)
genres_one_hot = copy[genres].join(copy['rating_percent']).dropna(subset = 'rating_percent')

# drop genres that have less than 50 games assigned to them and calculate the means
counts = copy[genres].sum()
valid_genres = counts[counts > 50].index
genre_means = (genres_one_hot[valid_genres].multiply(copy['rating_percent'], axis=0).sum() / counts[valid_genres]).sort_values()

# create the visualization and make it pretty
ratings_by_genre_vis = sns.barplot(x = genre_means.index, y = genre_means.values, palette='viridis')
ratings_by_genre_vis.set_xticklabels(ratings_by_genre_vis.get_xticklabels(), rotation=45, ha='right')
ratings_by_genre_vis.set_title('Positive Rating Ratio Given Game Genre') 
ratings_by_genre_vis.set(xlabel = 'Genre', ylabel = 'Average Positive Reviews Given Genre')
plt.show()
```

#### Results

Looking at the visualization, we can see that the lower-rated genres include 'Massively Multiplayer' and 'Gore'. MMOs tend to receive lower ratings due to long grinds and unfair monetary policies for their F2P players. Gorey games also tend to receive lower ratings due to being generally unappealing to wider audiences. ### Peak CCU (Concurrent players) vs Positive/Negative reviews

Next, we want to examine how peak CCU may affect the amount of positive and negative reviews a game may get. To do this, we will separate the Peak CCU into multiple bins, and from there compare the reviews to each bin to further understand how a game's peak popularity, or when people play the game the most, correlate with game reviews. This will allow us to get a more accurate reading from a bigger group of people, since at Peak CCU, it is more likely that the group of people who have played this game include those outside of the intended audience, giving a larger sample of players. 

```python
#Take needed columns into a separate dataframe
ccu_df = steam_games[["Peak CCU", "Positive", "Negative"]].copy()

#Remove any potential missing values
ccu_df = ccu_df.dropna()

# Keep only valid values within each column
ccu_df = ccu_df[
    (ccu_df["Peak CCU"] > 0) &
    (ccu_df["Positive"] >= 0) &
    (ccu_df["Negative"] >= 0)
]

# Create bins for Peak CCU
bins = [0, 10, 100, 1000, 10000, 100000, ccu_df["Peak CCU"].max()]
labels = ["0-10", "11-100", "101-1,000", "1,000-10,000", "10,000-100,000", "100,000+"]

ccu_df["CCU Range"] = pd.cut(
    ccu_df["Peak CCU"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

#Then, group by bins, and get the avg positive/negative reveiws
ccu_grouped = ccu_df.groupby("CCU Range")[["Positive", "Negative"]].mean()

#Format it correctly for seaborn
ccu_format = ccu_grouped.reset_index().melt(
    id_vars = "CCU Range",
    value_vars = ["Positive", "Negative"],
    var_name = "Review Type",
    value_name = "Average Reviews"
)

#Create plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=ccu_format,
    x = "CCU Range",
    y = "Average Reviews",
    hue = "Review Type"
)

#Label plot
plt.title("Average Positive and Negative Reviews by Peak CCU Range")
plt.xlabel("Peak CCU Range")
plt.ylabel("Average Number of Reviews")
plt.xticks(rotation=30)
plt.show()
```

This bar chart shows the average number of positive and negative reviews across different Peak CCU ranges. Note that as the Peak CCU increases, so does the number of positive and negative review count, showing that more popular games review significantly more reviews. This is expected, as there would be more opportunities to leave reviews. However, across all ranges, positive reviews seem to consistently surpass negative reviews, showing that the either people are more likely to leave positive reviews rather than negative, or overall tend to be more favorable towards the game. However, one important thing to note is how the lower Peak CCU ranges have a higher positive to negative ratio, whereas the popular games with over 100k in Peak CCU tend to have more negative reviews than other ranges in terms of percentages.

### Machine Learning: K-Nearest Neighbors

```python
from sklearn.compose import make_column_selector as selector
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
```

We will now try to see if there is a pattern amongst the related variables consistent enough to generate predictions of rating. First, we will assign labels to each game based on the % of their rating for the algorithm to predict. For example, if a game had 80% positive reviews, it would get a string label of '76-100% positive rating'. We will implement the K-Nearest Neighbors algorithm as we do not appear to have a high-dimensional problem with the number of examined features being small compared to our sample.

```python
# return a string label of the game's % of positive reviews, based on the number of entered positive and negative reviews, or no reviews if no reviews
def get_percent(pos,neg):
 # (below gives rounded percent to 2 decimal places)
    percent_pos = 0
    final_percent = 'thisshouldntshowup'
    
    if (pos + neg) > 0:
        percent_pos = round((pos / (pos + neg)), 2)
    else:
        final_percent = 'no reviews'
        return final_percent
    
    if (0 <= percent_pos <= 0.25):
        final_percent = '0-25% positive rating'
    elif (0.26 <= percent_pos <= 0.5):
        final_percent = '26-50% positive rating'
    elif (0.51 <= percent_pos <= 0.75):
        final_percent = '51-75% positive rating'
    elif (0.76 <= percent_pos <= 1):
        final_percent = '76-100% positive rating'
        
    return final_percent 

# form a new column on the steam games dataset with the string % of positive reviews
steam_games['Percent Positive Reviews'] = steam_games.apply(lambda x: get_percent(x.Positive, x.Negative), axis=1)
```

We then assign Predictors and the predicted Dependent Variable. We exclude obviously irrelevant variables from Predictors such as name and AppID which would give irrelevant noise, and must exclude median playtimes as they would be correlated with average playtimes and introduce a multicollinearity problem. 
Then, we form our train test split groups, stratifying based on our dependent variable as we have a small underrepresented number of games with lower %ge positive ratings, as reflected by our EDA above. 

```python
# excluding games that got no reviews, attempting to predict rating as a %ge range from other numerical and cat data
# picking only playtime average for 2 weeks and forever as both average and mean would be correlated
X = steam_games.drop(['Percent Positive Reviews','name','AppID', 'release_date','is_free','Price Range', 'Percentage of Positive', 'genres','categories', 'Median playtime two weeks','Median playtime forever'], axis=1)[steam_games['Percent Positive Reviews'] != 'no reviews']
y = steam_games[steam_games['Percent Positive Reviews'] != 'no reviews']['Percent Positive Reviews']

# stratify needed because of imbalanced dataset, very few low ratings 
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, test_size=0.5, random_state=42, stratify=y)
```

A Pipeline simplifies the process of preparing our predictors for analysis, as the one-hot encoder converts categorical data into numerical features that can be processed by the algorithm, while the standard scaler ensures that numerical variables with more extreme variations and range (eg. games having extreme differences in average playtime as compared to price in dollars generally being around a range) do not hold more weight in the model.

```python
num_col_selector = selector(dtype_exclude = object)
cat_col_selector = selector(dtype_include = object)

num_col = num_col_selector(X)
cat_col = cat_col_selector(X)


preprocessor = ColumnTransformer([ 
    ("one-hot-encoder", OneHotEncoder(), cat_col),
    ("standard_scaler", StandardScaler(), num_col)
])

steam_games_clf = Pipeline([ ('preprocessing', preprocessor), ('classifier',KNeighborsClassifier())])
```

We perform a grid search to refine our model and find the most optimal number of neighbors for our classification, given by the best_params attribute.

```python
param_grid = {
    'classifier__n_neighbors': [3,5,7,10,15]

}
best_model = GridSearchCV(steam_games_clf, param_grid, cv=5, verbose=1)

best_model.fit(X_train, y_train)
```

```python
best_model.best_params_ 
```

We generate an evaluation of Precision and Recall of the model's predicted test output of %positive ratings as compared to the true test set. We can see that precision and recall are substantially lower for games with 50% or below positive ratings, and this is likely due to the few numbers of games we have in the dataset with a percentage of ratings this low.

```python
steam_games_predicted_test_Y = best_model.predict(X_test)

class_report_test = classification_report(y_test, steam_games_predicted_test_Y)
print(class_report_test)

```
