# Research Question

**How do variables such as game genre, price, average length of playtime, and achievement number influence how Steam games are rated on the Steam shop globally since the platform's launch in 2003? Can these variables be used to predict how a Steam game might be rated?**

Note: All the variables are attributes of games listed in the Steam storefront page for a game.
Steam game genre refers to the available platforms and number of players it is targeting, price refers to the amount of currency in dollars required to purchase (including free games), average length of playtime refers to the average amount of time in hours it takes a player to reach an ending of the game for the first time, and achievement number refers to how many achievements can be attained by meeting in-game requirements for a game.
We are restricting our population to games on Steam, and thus only reviewing ratings left on the Steam platform. 
We are interested in sampling across a time period of 26 years and investigating whether there are any significant patterns in the aforementioned variables. 



## Background and Prior Work

Steam, a digital distribution service, lets its users buy video games from its digital store with a catalog of thousands of publishers, as well as install into and play video games from the users’ own digital library.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Important attributes of the games, such as their genre, price, number of game achievements are displayed on their store pages, allowing interested players to quickly get a feel for the game they may want to purchase.

### Which genre is best?

We wanted to explore the effect that genre had on how well a game performed in the Steam shop. To achieve this, we used the ratio of positive reviews to total reviews as our test statistic and used one-hot encoding to compensate for the fact that each game has multiple genres attached. One challenge we encountered while creating this EDA was that the 'rating_percent' column is originally a string containing the code for a Python list that held the genres. To solve this, we had to import the 'ast' package. Additionally, some genres had artificially inflated scores due to having very few entries, so we dropped them. 

```python
# create a copy of the steam_games dataframe for manipulation
copy = steam_games.copy()

# add a column 'rating_percent' that describes the ratio of positive reviews to total reviews  
copy['rating_percent'] = copy['Positive'] / (copy['Positive'] + copy['Negative'])
# convert the 'genres' column from strings to lists 
copy['genres'] = copy['genres'].apply(lambda x: ast.literal_eval(x))

# get all the unique genres from the genres column
genres = pd.Series([i for lst in copy['genres'] for i in lst]).unique()

# create a new one-hot column for every unique genre in 'genres'
for genre in genres:
    copy[genre] = copy['genres'].apply(lambda x: genre in x)
genres_one_hot = copy[genres].join(copy['rating_percent']).dropna(subset = 'rating_percent')

# drop genres that have less than 50 games assigned to them and calculate the means
counts = copy[genres].sum()
valid_genres = counts[counts > 50].index
genre_means = (genres_one_hot[valid_genres].multiply(copy['rating_percent'], axis=0).sum() / counts[valid_genres]).sort_values()

# create the visualization and make it pretty
ratings_by_genre_vis = sns.barplot(x = genre_means.index, y = genre_means.values, palette='viridis')
ratings_by_genre_vis.set_xticklabels(ratings_by_genre_vis.get_xticklabels(), rotation=45, ha='right')
ratings_by_genre_vis.set_title('Positive Rating Ratio Given Game Genre') 
ratings_by_genre_vis.set(xlabel = 'Genre', ylabel = 'Average Positive Reviews Given Genre')
plt.show()
```

### Results

Looking at the visualization, we can see that the lower-rated genres include 'Massively Multiplayer' and 'Gore'. MMOs tend to receive lower ratings due to long grinds and unfair monetary policies for their F2P players. Gorey games also tend to receive lower ratings due to being generally unappealing to wider audiences. On the other hand, 'Indie' games are generally more well-received since they are often more passionate projects compared to AAA games, while 'Action' games are more engaging, creating a sharp contrast with the lower ratings given to the 'Massively Multiplayer' genre. As a whole, genres tend to have a significant effect on the rating of a game.
